```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_marks.csv")

plt.figure(figsize=(8,6))
bars = plt.bar(df["Subject"], df["Average_Marks"], color=["skyblue", "lightgreen", "salmon", "orange", "violet"])

plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Average Marks", fontsize=12)
plt.title("Average Marks of Students in Different Subjects", fontsize=14, fontweight="bold")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             str(round(bar.get_height(),1)), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_category.csv")

plt.figure(figsize=(8,6))
bars = plt.bar(df["Category"], df["Sales"],
               color=["skyblue", "lightgreen", "salmon", "orange", "violet", "gold"])

plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Sales (in Units or $)", fontsize=12)
plt.title("Sales Distribution Across Product Categories", fontsize=14, fontweight="bold")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             str(bar.get_height()), ha='center', va='bottom', fontsize=10)

plt.show()
```
